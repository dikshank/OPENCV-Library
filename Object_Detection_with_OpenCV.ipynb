```python
methods = ['cv2.TM_CCOEFF','cv2.TM_CCOEFF_NORMED','cv2.TM_CCORR','cv2.TM_CCORR_NORMED','cv2.TM_SQDIFF','cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:

  #  create a copy of image
  full_copy = full.copy()
  
  method = eval(m)


  # template matching

  res = cv2.matchTemplate(full_copy,sub_img,method)

  min_val,max_value,min_loc,max_loc = cv2.minMaxLoc(res)

  if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
    top_left = min_loc

  else:
    top_left = max_loc

  height,width,channels = sub_img.shape

  bottom_right = (top_left[0]*width,top_left[1]+height)

  cv2.rectangle(full_copy,top_left,bottom_right,(255,0,0),10)

  # plot and show the images

  plt.subplot(121)
  plt.imshow(res)
  plt.title('HEATMAP OF TEMPLATE MATCHING')

  plt.subplot(122)
  plt.imshow(full_copy)
  plt.title('DETECTION OF TEMPLATE')

  # Title with the method used
  plt.suptitle(m)
  plt.show()

  print('\n')
  print('\n')
```

# EDGE DETECTION

### CANNY EDGE DETECTION

```python
img = cv2.imread('/content/drive/MyDrive/OBJECT_DECTECTION_WITH_OPENCV/cat-551554__340.jpg')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img,threshold1=127,threshold2=127)
```

```python
plt.imshow(edges)
```

```python
med_val = np.median(img)
```

```python
med_val
```

```python
#lower threshold to either 0 or 70% of the median value whichever is grater

lower = int(max(0,0.7*med_val))

#lower threshold to either 130% of the median value or the max 255 whichever is lower

upper = int(min(255,1.3*med_val))
```

```python
edges = cv2.Canny(image=img,threshold1=lower,threshold2=upper+100)
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img,ksize=(7,7))
```

```python
edges = cv2.Canny(image=blurred_img,threshold1=lower,threshold2=upper+50)
plt.imshow(edges)
```
